WHR2018_6.ipynb is a Jupyter Notebook written in Python for Analysis of World Happiness Report 2018 (Kaggle.com, /kaggle/input/world-happiness-report/2018.csv) using Pandas, Seaborn, Matplotlib.pyplot. 

Includes EDA, correlation matrix, and analysis of 6 countries.

```python
import pandas as pd
whr2018=pd.read_csv("2018.csv")
```

```python
whr2018.info()
```

```python

Place = pd.DataFrame(whr2018[['Country or region', 'Overall rank']])
whr2018.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
whr2018sm=whr2018.drop(columns =['Overall rank'  ,'Country or region', 'Score'])
corr=whr2018sm.corr()
sns.heatmap(corr, annot = True, fmt = '.2f', cmap ='coolwarm' , linewidth = .5)
plt.title('Correlation of Features for World Happiness 2018')
plt.show()
```

```python
import matplotlib.pyplot as plt
print(whr2018.head(3))
print(whr2018.tail(3))
whr6 = pd.concat([whr2018.head(3), whr2018.tail(3)])

sns.barplot( y = 'Country or region', x= 'Score', data = whr6, color = 'grey')
plt.title('Bar Chart Countries by Overall Hapiness')
plt.ylabel('Countries')
plt.xlabel('Score')
plt.show()
```
